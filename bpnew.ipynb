```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Features
x = np.array([
    [-1, -1],
    [-1, 1],
    [1, -1],
    [1, 1]
])

# Target
y = np.array([-1, 1, 1, -1])
alpha = 0.001
epoch = 1400
```

```python
# Initialization of weights (input layer)
a = -0.5
b = 0.5
# [
#  [w10, w11, w12],
#  [w20, w21, w22]
# ]
wij = (b - a) * np.random.random_sample((2,3)) + a

# Bias
bias_in = np.array([1 , 1])
wij
```

```python
# Initialization of weights ( output layer)
# [
#   [w10, w11, w12 ]
# ]
wjk = ( b - a ) * np.random.random_sample((1,3)) + a

# Bias
bias_out = np.array([1])
wjk
```

```python
def sigmoid(mat):
    return ((1.0) / (1 + np.exp(-mat)))
```

```python
def display(itr):
    print(f'\n----------Iteration {itr + 1} -------------------\n')
    print(f' Wij :  \n{wij} \n')
    print(f'Bias in : {bias_in}\n')

    print(f' Wjk :  \n{wjk}\n')
    print(f'Bias out : {bias_out}\n')
```

```python
print("Initialization......")
display(0)
```

```python
for epo in range(epoch):
    print(f'------------Epoch: {epo + 1}-------------')
    for i in range(x.shape[0]):

        # Required output
        tk = y[i]

        # Preparing for input
        xin = np.array([[bias_in[0], x[i][0], x[i][1]]])
        xin = xin.transpose()
        # print(f'xin : {xin}')

        # Zin_j       
        Zin = wij @ xin
        
        # Get Zj
        Zj = sigmoid(Zin)
        # print(f'Zin : {Zin}')
        # print(f'Zj: {Zj}')
        
        Z = np.array([[bias_out[0], Zj[0][0], Zj[1][0]]]) 
        Z = Z.transpose() # For input into output neuron

        # Yin_k
        yin_k = wjk @ Z
        # print(f'Yin_k : {yin_k}')
        yk = sigmoid(yin_k)
        # print(f'Yk : {yk}')

        # del_k
        del_k = (tk - yk) * yk * (1 - yk)
        err_ho.append(del_k[0][0])
        # print(f'del_k : {del_k}')

        # del_wjk
        del_wjk = alpha * ( del_k @ (Z.transpose()))
        # print(f'del_wjk : {del_wjk}')

        # del_in_j
        t_wjk = np.array([[wjk[0][1], wjk[0][2]]]) # improve
        t_wjk = t_wjk.transpose()
        del_inj = t_wjk @ del_k
        # print(f'del_inj : {del_inj}')
        
        del_j = del_inj * Zin * ( 1 - Zin)
        # print(f'del_j : {del_j}')

        #del_Wij
        del_wij = alpha * (del_j @ (xin.transpose()))
        #print(f'del_wij : {del_wij}')
                           

        # Update
        wij += del_wij
        wjk += del_wjk
        
        display(i)

```
